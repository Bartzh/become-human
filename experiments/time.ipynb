```python
import sys
import os

# 添加父目录到路径
sys.path.insert(0, os.path.abspath('..'))

from datetime import datetime
from ipywidgets import Text
from IPython.display import display
from become_human.times import TimestampUs, utcnow

now = utcnow()
datetime_textbox: Text = Text(
    description='Datetime:',
    value=str(now),
    style={'description_width': 'initial'},
    continuous_update=False)
timestampus_textbox: Text = Text(
    description='TimestampUs:',
    value=str(int(TimestampUs(now))),
    style={'description_width': 'initial'},
    continuous_update=False)
def update_timestampus_textbox(change):
    timestampus_textbox.value = str(int(TimestampUs(datetime.fromisoformat(change['new']))))
datetime_textbox.observe(update_timestampus_textbox, names='value')
def update_datetime_textbox(change):
    datetime_textbox.value = str(TimestampUs(int(change['new'])).to_datetime())
timestampus_textbox.observe(update_datetime_textbox, names='value')
display(datetime_textbox)
display(timestampus_textbox)
```
