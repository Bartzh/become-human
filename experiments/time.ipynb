```python
import sys
import os

# 添加父目录到路径
sys.path.insert(0, os.path.abspath('..'))

from datetime import datetime
from ipywidgets import Text
from IPython.display import display
from become_human.times import datetime_to_seconds, seconds_to_datetime, nowtz

now = nowtz()
datetime_textbox: Text = Text(
    description='Datetime:',
    value=str(now),
    style={'description_width': 'initial'},)
timeseconds_textbox: Text = Text(
    description='Time seconds:',
    value=str(datetime_to_seconds(now)),
    style={'description_width': 'initial'},)
def update_timeseconds_textbox(sender):
    timeseconds_textbox.value = str(datetime_to_seconds(datetime.fromisoformat(sender.value)))
datetime_textbox.on_submit(update_timeseconds_textbox)
def update_datetime_textbox(sender):
    datetime_textbox.value = str(seconds_to_datetime(float(sender.value)))
timeseconds_textbox.on_submit(update_datetime_textbox)
display(datetime_textbox)
display(timeseconds_textbox)
```
